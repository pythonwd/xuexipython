```python
from requests_html import HTMLSession

import urllib.request 
import time 
session=HTMLSession() 
urls = ['http://www.win4000.com/zt/mao_{}.html'.format(str(i)) for i in range(1, 3)] 
# url='http://www.win4000.com/zt/mao_1.html' 
for url in urls: 
    r=session.get(url) 
    time.sleep(1) 
    hrefs=r.html.find('div.tab_box>div>ul.clearfix>li>a>img') 
    hrefs2 = r.html.find('div.tab_box>div>ul.clearfix>li>a') 
    for href2 in hrefs2[:-5]: 
        # print(href2.find('img')) 
        name=href2.text 
        print(name)
        for i in href2.find('img'): 
            print(i.attrs['data-original']) 
            #这就是图片下载地址，str类型 
            urllib.request.urlretrieve(i.attrs['data-original'], '/home/wangd/图片/pachong\%s.jpg' %name)
```
