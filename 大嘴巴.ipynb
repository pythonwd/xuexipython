```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#import jieba  #如果是中文一定要这个来分词
from os import path
import collections
import pandas as pd
df = pd.read_excel("/home/wangd/xl.xlsx",sheetname=4)
df = df.loc[df["业 务 员"].isin(["王东（大）"])]
nr = df["存货名称"]
# with open("/home/wangd/思源.text") as f:
#     nr=f.read()
#text = ' '.join(jieba.cut(nr))
word_counts = collections.Counter(nr) # 对分词做词频统计
# word_counts_top20 = word_counts.most_common(20) # 获取前10最高频的词
mk = plt.imread('/home/wangd/guibao6.jpg') 
wordcloud = WordCloud(font_path="/home/wangd/msyh/msyh.ttf",mask=mk,background_color='white',
                     width=1000, height=1000).generate_from_frequencies(word_counts) #字体路径、图片路径、背景颜色、画布高宽
%pylab inline
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#print(word_counts)
```
